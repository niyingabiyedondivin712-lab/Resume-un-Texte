```python
import os
from groq import Groq
from dotenv import load_dotenv

load_dotenv()
cle = os.getenv("Groq_API_KEY")
client = Groq(api_key=cle)

def Resume(text):

    respond = client.chat.completions.create(
        model='openai/gpt-oss-20b',
        messages=[
            {'role':'system',
             'content':'fais une resume de texte de ce que je vous envoie puis par defaut commence par 5 de resume mais une utilisateur a deja entre les ligne tu doit respecte les ligne'},
             {'role':'user',
              'content':text}
        ]
    )

    return respond.choices[0].message.content


texte = input("Entre un texte pour te faire une resume")

print(Resume(texte))
```
